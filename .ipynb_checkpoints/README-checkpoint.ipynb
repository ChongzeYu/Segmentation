# Sample Submission
This notebook presents a sample submission for the [Outdoor Semantic Segmentation Challenge](https://unstructured-scene-understanding.com/challenge.html). The submission on Codalab requires you to submit your prediction maps in a zip file containing your predictions as MAT-Files. The code on how to create the submission zip is available in this notebook.

Here we use [Fast-SCNN](https://arxiv.org/pdf/1902.04502.pdf) trained on the [TAS500 dataset](https://www.mucar3.de/icpr2020-tas500/). The model was trained for 150 epochs and no data augmentations were applied.

This notebook is also able to run the model inference on machines without a GPU, by just using the CPU.

You can install al necessary Python packages for this notebook by executing `pip install -r requirements.txt` in this directory.

```python
import os
import os.path as osp
import shutil
import sys
import zipfile

from google_drive_downloader import GoogleDriveDownloader as gdd
import imageio
from matplotlib.lines import Line2D
from matplotlib.patches import Patch
import matplotlib.pyplot as plt
import numpy as np
from scipy import io
import torch
import torch.backends.cudnn as cudnn
import torch.nn as nn
import torch.nn.functional as F


from model.model_builder import build_model
from viz.colormap import get_class_colors, get_class_names, visualize_prediction # helper scripts for visualization

%matplotlib inline
```

```python
model="FastSCNN"
weights=osp.join('model','model_150.pth')
classes=23
gpu="1"  # if you have a GPU available, then set this ID accordingly

class LearningModel:
    def __init__(self):
        self.model = build_model(model, num_classes=classes)
        
        if not torch.cuda.is_available():
            checkpoint = torch.load(weights, map_location=lambda storage, loc: storage)
            self.model.load_state_dict(checkpoint['model'])
        else:
            checkpoint = torch.load(weights)
            self.model.load_state_dict(checkpoint['model'])
            
            os.environ["CUDA_VISIBLE_DEVICES"] = gpu
            self.model = self.model.cuda()
            cudnn.benchmark = True

    def predict_img(self, img, softmax=False, only_max=False, consolidate_classes=False):
        img_tensor = torch.from_numpy(img).float()[None,:,:,:].permute(0,3,1,2)
        self.model.eval()
                
        with torch.no_grad():
            if torch.cuda.is_available():
                input_var = img_tensor.cuda()
            else:
                input_var = img_tensor
                
            output = self.model(input_var) # (batch,channel,height,width)
            output = F.softmax(output, dim=1)
                
        output = output.cpu().data[0].numpy()
        output = output.transpose(1, 2, 0) # (height,width,channel)
        output = np.asarray(np.argmax(output, axis=2), dtype=np.uint8)
        
        return output
    

fast_scnn = LearningModel()
```

We run the inference of our Fast-SCNN model on the test images of the TAS500 dataset. The test images are part of the TAS500 dataset download, but they are also separately available [here](https://drive.google.com/file/d/1IL3jAcsqxLPANPi8Mo_eBlqaYN9zz0r1/view?usp=sharing) on Google Drive (the download script requires the [**googledrivedownloader**](https://pypi.org/project/googledrivedownloader/) package to be installed).


```python
google_drive_id = '1IL3jAcsqxLPANPi8Mo_eBlqaYN9zz0r1'
test_images_dest_path = './test_input_data/codalab_test_input_data.zip'

gdd.download_file_from_google_drive(file_id=google_drive_id,
                                    dest_path=test_images_dest_path,
                                    unzip=True)

# remove the .zip file after the download
if osp.exists(test_images_dest_path):
    os.remove(test_images_dest_path)
```

```python
# predict the semantic map for every image in the test set
# and store the predictions in a dictionary of the following form
# {
#   '1527246337974638592.png' : [[12 12 12 ...  5  5  5]
#                                [12 12 12 ...  5  5  5]
#                                [12 12 12 ...  5  5  5]
#                                ...
#                                [ 6  6  6 ...  4  4  4]
#                                [ 6  6  6 ...  4  4  4]
#                                [ 6  6  6 ...  4  4  4]],
# ...
# }
test_data_dir = 'test_input_data' # place all images of the test set into a dictionary
submission = {}

fast_scnn.model.train(False)

for img_name in os.listdir(test_data_dir):
    img = imageio.imread(osp.join(test_data_dir, img_name))
    input_data = np.array(img[:,:,:3])
    pred = fast_scnn.predict_img(input_data)
    submission[img_name] = np.array(pred)
```

```python
# visualize a prediction in color
scene_id = '1567610833046423624.png'

img = imageio.imread(osp.join(test_data_dir, scene_id))
color_pred = visualize_prediction(submission[scene_id])

fig = plt.figure(figsize=(15,11))

ax = fig.add_subplot(2,1,1)
ax.set_title('Input Image')
ax.imshow(img)
ax.set_axis_off()

ax = fig.add_subplot(2,1,2)
ax.set_title('Prediction')
ax.imshow(color_pred)
ax.set_axis_off()

labels = get_class_names(np.unique(submission[scene_id]))
        
colors = get_class_colors(np.unique(submission[scene_id]))
legend_elements = [Line2D([0], [0], linewidth=0,
                    marker='s', color='w',
                    label=pair[0], 
                    markerfacecolor=list(map(lambda x: x / 255.,pair[1])),
                    markersize=9) for pair in zip(labels,colors)]

legend = ax.legend( loc="lower right", bbox_to_anchor=(0.75, 0, 0.25, 0.25), # (x,y,w,h)
                   
                    handles=legend_elements,
                    mode=None,
                    ncol=len(labels)//2,
                    title="Semantic Classes",
                    title_fontsize="x-large",
                    fontsize="medium",
                    columnspacing=1.5,
                    edgecolor="black")
```

If we are happy with our results on the test set, then we can pack the submission as a zipped list of .mat files containing the class ID for each pixel:

```python
# create the zip file for submission
with zipfile.ZipFile('test_pred.zip', 'w') as zip_handle:
    if not osp.exists('tmp'):
        os.mkdir('tmp')
    for key in submission.keys():
        io.savemat('tmp/{}.mat'.format(key[:-4]), {'data': submission[key].astype(np.uint8)}, do_compression=True) 
        zip_handle.write('tmp/{}.mat'.format(key[:-4]), arcname=key[:-4] + '.mat')
    shutil.rmtree('tmp')
```
